# Monuments of Madrid

```python
import folium
import pandas as pd
from urllib.request import Request, urlopen
```

```python
#Request data

req = Request('https://datos.madrid.es/egob/catalogo/300356-0-monumentos-ciudad-madrid.csv')
req.add_header('User-Agent', 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0')
content = urlopen(req)



df = pd.read_csv(content,encoding='ISO-8859-1',sep=';')

```

```python
#df.to_csv('df_output.csv', encoding='utf-8')
#df2=pd.read_csv('df_output.csv', encoding='utf-8',sep=',')
#df2.head()

df.head()
```

```python
import numpy as np

#Drop NaN values from d
df1 = df[pd.notnull(df['LATITUD'])]
df1.head()
```

```python
#find index of nan 
index = df1['LONGITUD'].index[df1['LONGITUD'].apply(np.isnan)]

#check for NaN on dataset

index
```

```python
df1 = df1.replace('í', 'i') 
df1.head()
```

```python
from folium import plugins


lat=40.4167754
long=-3.7037902

# let's start again with a clean copy of the map of San Francisco
madrid_map = folium.Map(location = [lat, long], zoom_start = 11)

# instantiate a mark cluster object for the incidents in the dataframe
monuments = plugins.MarkerCluster().add_to(madrid_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df1['LATITUD'], df1['LONGITUD'], df1['NOMBRE']):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(monuments)

# display map
madrid_map
```
